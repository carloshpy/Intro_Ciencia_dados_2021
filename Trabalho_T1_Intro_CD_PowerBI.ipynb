```python
df_order_pay.payment_type.value_counts()
```

```python
df_order_pay[df_order_pay.payment_value>0].groupby('payment_type').payment_value.mean().sort_values(ascending= False).plot(kind = "bar")
```

```python
df_order_pay[df_order_pay.payment_value>0].groupby('payment_type').order_id.nunique().sort_values(ascending= False).plot(kind = "bar")
```

```python
df_order_pay[df_order_pay.payment_installments>0].groupby('payment_installments').payment_value.mean().sort_values(ascending= False).plot(kind = "bar")
```
